**선 그래프 그리기 응용**

```
from matplotlib import pyplot as plt

x = ['유재석','박명수','하하','정형돈']
y_1 = [85, 100, 95, 88]
y_2 = [79, 97, 87, 92]
plt.plot(x, y_1, label='수학', color='red', linestyle='-.',marker='o') # plt.plot() 메소드로 선 그래프 그리기(첫번째 인자는 x축, 두번째 인자는 y축 값), 선 스타일(색상, 선 종류, 마커)
plt.plot(x, y_2, label='영어' ) # plt.plot() 메소드로 선 그래프 그리기(첫번째 인자는 x축, 두번째 인자는 y축 값)
plt.legend(loc='lower right') # 범례 표시(우측 하단) #upper/center/lower + left/center/right
plt.title('수학/영어 성적') # 그래프 제목 설정
plt.xlabel('이름') #X축 레이블 이름 설정
plt.ylabel('점수') #Y축 레이블 이름 설정
plt.ylim(70,105) # Y축의 범위 설정
plt.grid(True) # 그리드 출력
plt.show() # 그래프를 화면에 출력
```

**다수의 그래프 그리기**

```
import matplotlib.pyplot as plt

x = ['월','화','수','목 ','금','토','일']
y1 = [2.5, 1.5, 0.7, 2.7, 2.3, 2.9, 1.2]
y2 = [1.3, 2.5, 1.1, 2.1, 1.7, 2.4, 2.7]
y3 = [2.6, 1.4, 4.1, 1.1, 2.4, 1.2, 3.1]
y4 = [2.4, 3.1, 2.6, 1.5, 2.9, 1.9, 1.6]

fig, axs = plt.subplots(nrows=2, 
                        ncols=2, figsize=(12, 7), sharex=True, sharey=True) # (12,7) 크기의 2x2 subplot 선언

final = axs[0][0] # subplot(0,0)을 final에 대입
final.scatter(x, y1) #산점도 그리기
final.set_title('1월') # 그래프 제목 설정

final = axs[0][1] # subplot(0,1)을 final에 대입
final.bar(x, y2, color='green') #막대 그래프 그리기, 그래프 색상 : 녹색
final.set_title('2월') # 그래프 제목 설정

final = axs[1][0] # subplot(1,0)을 final에 대입
final.bar(x, y3, width=0.3, color='blue') #막대 그래프 그리기, 그래프 너비 : 0.3, 그래프 색상 : 파랑
final.set_title('3월') # 그래프 제목 설정

final = axs[1][1] # subplot(1,1)을 final에 대입
final.plot(x, y4, color='red', linestyle='--', marker='x') #선 그래프 그리기, 스타일 지정, 마커 지정
final.set_title('4월') # 그래프 제목 설정

fig.suptitle('요일별 평균 운동시간') # fig 그래프 제목 설정
plt.show() # 그래프를 화면에 출력
```

```
import matplotlib.pyplot as plt
import seaborn as sns

tips = sns.load_dataset("tips") # catplot을 활용하여 범주형 변수와 수(치)형 변수간 관계 시각화
sns.catplot(x="day",y="total_bill", data=tips)  # x는 x축 제목, y는 y축 제목, data는 그래프에 들어가는 데이터
sns.catplot(x="day",y="total_bill", hue="smoker", kind="bar",data=tips) # x는 x축 제목, y는 y축 제목, hue는 hue semantic 제목, data는 그래프에 들어가는 데이터, kind는 그래프 종류
sns.catplot(x="day", y="total_bill", hue="smoker", kind="violin", data=tips) # x는 x축 제목, y는 y축 제목, hue는 hue semantic 제목, data는 그래프에 들어가는 데이터, kind는 그래프 종류
plt.figure(figsize=(8,6)) # figure 크기를 8,6으로 설정
```

# **3. Scikit-learn을 활용한 머신러닝**

*   항목 추가
*   항목 추가



```
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
import pandas as pd
from sklearn.metrics import accuracy_score
import seaborn as sns

iris = load_iris() # 붓꽃 데이터 셋 로딩
#print(iris)

iris_data = iris.data # iris.data : Iris 데이터셋에서 feature 데이터
iris_label = iris.target # iris.target : Iris 데이터셋에서 레이블 데이터
#print('iris data 값: ', iris_data[:5])
#print('iris target의 값: ', iris_label)
```

```
iris_df = pd.DataFrame(data=iris_data, columns=iris.feature_names) # Iris 데이터셋을 DataFrame으로 변환
iris_df ['label'] = iris.target # iris_df에 iris target data 추가
#iris_df.head()

sns.pairplot(iris_df, x_vars=["sepal length (cm)"],  # 첫 번째 인자: 데이터, x_vars: x 축 변수 설정
             y_vars=["sepal width (cm)"], hue="label", height=5) # y_vars: y축 변수 설정, hue: label 제목 설정, height: plot 크기 설정
```

```
X_train, X_test, y_train, y_test = train_test_split(iris_data, iris_label, test_size=0.2) # 데이터셋 분리

dt_clf = DecisionTreeClassifier() # DecisionTreeClassifier 객체 생성
knn_clf = KNeighborsClassifier() # KNeighborsClassifier 객체 생성

dt_clf.fit(X_train, y_train) # 모델 학습 수행
knn_clf.fit(X_train, y_train) # 모델 학습 수행

y1_pred = dt_clf.predict(X_test) # 예측
y2_pred = knn_clf.predict(X_test) # 예측

print('DecisionTree 예측정확도 : {0:.4f}'.format(accuracy_score(y_test, y1_pred))) # 평가
print('k-NN 예측정확도 : {0: .4f}'.format(accuracy_score(y_test, y2_pred))) # 평가
```
